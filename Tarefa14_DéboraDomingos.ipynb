**1)** Para este item você deverá usar o arquivo com dados de força e torque medidos no chão enquanto uma pessoa tenta ficar em pé durante um minuto.

*   Faça o gráfico do torque na direção y (My [Nm]) em função do tempo (Time [s]).
*   Mostre na tela a média e o desvio-padrão do torque na direção y.
    



```python
import pandas as pd
import numpy as np
Torque = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/balance.csv')
Torque
Torque.plot('Time[s]', 'My[Nm]', kind='line', figsize=(10,8), grid=True, ylabel='Torque',
       title='Torque na direção y em função do tempo', fontsize=15);
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mediatorque = Torque['My[Nm]'].mean()
print(f' A média do torque na direção y é {round(mediatorque,0)}')

stdX = Torque['My[Nm]'].std()
plt.figure()
plt.plot(Torque['My[Nm]'], marker = 'o', linestyle = '', label='Torque', color = 'black')
plt.axhline(Torque['My[Nm]'].mean(), marker = '',
         linestyle = '-', color = 'red', label='média')
plt.axhline(Torque['My[Nm]'].mean() + stdX, marker = '', 
         linestyle = '--', color = 'blue', label='desvio-padrão')
plt.axhline(Torque['My[Nm]'].mean() - stdX, marker = '', 
         linestyle = '--', color = 'blue')
plt.legend()
plt.show()
```

*     Faça o histograma da inflação mensal nos meses de março.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

inflacao = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv')
medianainflacao = inflacao.query("Mês  == 3")['Inflação'].median()
mediainflacao = inflacao.query("Mês  == 3")['Inflação'].mean()
março = inflacao.query("Mês  == 3")['Inflação']
desvio = inflacao.query("Mês  == 3")['Inflação'].std()
print(f' A mediana da inflação mensal nos meses de março foi de {round(medianainflacao,2)}')
print(f' A média da inflação mensal nos meses de março foi de {round(mediainflacao,2)}')
print(f' O desvio padrão da inflação mensal nos meses de março foi de {round(desvio,2)}')

plt.figure()
plt.hist([inflacao.query("Mês == 3")['Inflação']])
plt.xlabel('Inflação')
plt.grid()
plt.show()

```

Faça isso para

```python
N = 100
M = 1000
O = 10000
P = 100000
rng = np.random.default_rng()

xn = rng.uniform(size=N)
xm = rng.uniform(size=M)
xo = rng.uniform(size=O)
xp = rng.uniform(size=P)

x1 = xn.mean()
x2 = xm.mean()
x3 = xo.mean()
x4 = xp.mean()

plt.figure()
plt.hist(xn)
plt.xlabel(f"RNG size {N}")
plt.grid()
plt.show()

print(f' Média dos valores gerados : {round(x1,2)}')

plt.figure()
plt.hist(xn)
plt.xlabel(f"RNG size {M}")
plt.grid()
plt.show()

print(f' Média dos valores gerados : {round(x2,2)}')

plt.figure()
plt.hist(xn)
plt.xlabel(f"RNG size {O}")
plt.grid()
plt.show()

print(f' Média dos valores gerados : {round(x3,2)}')

plt.figure()
plt.hist(xn)
plt.xlabel(f"RNG size {P}")
plt.grid()
plt.show()

print(f' Média dos valores gerados : {round(x4,2)}')
```
