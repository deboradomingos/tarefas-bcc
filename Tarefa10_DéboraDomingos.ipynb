```python
#Importando o Pandas
import pandas as pd

#Importando o Numpy
import numpy as np

#importando o matplotlib
import matplotlib.pyplot as plt

#importando o datetime
import datetime
```

```python
#Exercicio 1

#Importando a base
base = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')
```

```python
base.isnull().sum()
```

```python
base.loc[:, "Venceu Mandante"] = ""
base.head(3)
```

```python
base['Venceu Mandante'] = np.where(base['Placar do Mandante'] > base['Placar do Visitante'],'Venceu','Perdeu')
base.head()
```

```python
#Cria coluna auxiliar
base.loc[:, "Validador"] = ""
```

```python
#Cria diferença de gols para saber se venceu, perdeu ou empatou
base['Validador'] = base['Placar do Mandante'] - base['Placar do Visitante']
base.head()
```

```python
#Agrupamento para saber se ganhou, perdeu ou empatou:
#Valores = 0, entao empatou
#Valores menores que 0, perdeu
#Valores maiores que 0, ganhou.
agrupamento = base.groupby('Validador')['Validador'].count()
agrupamento
```

```python
#RESPOSTA:

#Empate (não Perdeu): 110
#Ganhou: 202
#Perdeu: 68

#% DE VITORIAS: 53%
#% DE NÃO DERROTAS: 28,9%
```

```python
#Exercicio 3

data = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/netflix.csv')

#tratamento 

data['Categoria'] = 'Filme'  
data['Categoria'][data['Title'].str.contains(": Temporada|: Stranger|: Parte")] = 'Série'  
data['Programa'] = data['Title']  
data[['Programa','Episódio']] = data[data['Categoria']=='Série']['Title'].str.split(pat = ': Temporada|: Stranger Things|: Parte', expand = True, n = 1)   
data.loc[data['Categoria']=='Filme', 'Programa'] = data.loc[data['Categoria']=='Filme', 'Title']  
data = data.drop(columns = 'Title')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
#analisando a estrutura dos dados
data.info()
```

```python
#agrupamento para saber as series mais assistidas
agrupamento2 = data.groupby('Programa')['Programa'].count().sort_values(ascending=False)
print('Esses são o top 10 series assistidos:', agrupamento2.head(10))
```

```python
#agrupamento para saber qual mes foi mais assistido.
#para isso, convertemos o campo date para um campo "date time"
data['Date'] = pd.to_datetime(data['Date'])
data['Mes'] = data['Date'].dt.month
data['Ano'] = data['Date'].dt.year

agrupamento3 = data.groupby('Mes')['Mes'].count().sort_values(ascending=False)
print('O mes com maior numero de series assistas é:' ,agrupamento3.head(1))
```
